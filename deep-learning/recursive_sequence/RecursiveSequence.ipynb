```python
s1 = 0.0001
w = 4
n = 50
```

```python
sequence = [s1]
for i in range(n-1):
    s_prev = sequence[-1]
    sequence.append(w*s_prev - w*s_prev**2)
```

```python
import matplotlib.pyplot as plt
plt.plot(sequence, 'r-')
plt.plot(sequence, 'bo')
plt.xlabel('steps')
plt.ylabel('value')
plt.show()
```
